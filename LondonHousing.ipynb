```python
#import pandas and numpyt
import pandas as pd
import numpy as np

# Load the pyplot collection of functions from matplotlib, as plt 
import matplotlib.pyplot as plt
```

```python
#loading the data
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

prop = pd.read_excel(url_LondonHousePrices, sheet_name = 'Average price', index_col = None)
```

```python
#exploring the data set
prop.shape
prop.head()
```

```python
# transpose of dataframe
prop_T = prop.T
prop_T.head()
```

```python
prop_T.index
```

```python
prop_T = prop_T.reset_index()
prop_T.index
```

```python
# print first row
prop_T.iloc[[0]]
```

```python
#assign the values of the first row to column heading
prop_T.columns = prop_T.iloc[0]

#drop row at index 0
prop_T = prop_T.drop(0)
prop_T.head()
```

```python
prop_T = prop_T.rename(columns = {'Unnamed: 0':'London Borough', pd.NaT:"ID"})
prop_T.head()
```

```python
prop_T.columns
```

```python
clean_prop = pd.melt(prop_T, id_vars = ['London Borough', 'ID'], var_name = 'Date', value_name = 'Average Price')
clean_prop.head()
```

```python
clean_prop.dtypes
```

```python
clean_prop['Average Price'] = pd.to_numeric(clean_prop['Average Price'])
clean_prop.dtypes
```

```python
clean_prop.count()
```

```python
clean_prop['London Borough'].unique()
```

```python
clean_prop[clean_prop['London Borough'] == 'Unnamed: 34'].head()
```

```python
clean_prop[clean_prop['London Borough'] == 'Unnamed: 37'].head()
```

```python
clean_prop[clean_prop['ID'].isna()]
```

```python
# getting rid of NA values
NaNFreeDF1 = clean_prop[clean_prop['Average Price'].notna()]
NaNFreeDF1.head(48)
```

```python
NaNFreeDF1.count()
```

```python
# drop all null values
NaNFreeDF2 = clean_prop.dropna()
NaNFreeDF2.head(48)
```

```python
NaNFreeDF2['London Borough'].unique()
```

```python
print(clean_prop.shape)
print(NaNFreeDF1.shape)
print(NaNFreeDF2.shape)
```

```python
# getting rid of nonboroughs
nonBoroughs = ['Inner London', 'Outer London', 'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER',
               'EAST MIDLANDS', 'WEST MIDLANDS', 'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
               'SOUTH WEST', 'England']
NaNFreeDF2 = NaNFreeDF2[~NaNFreeDF2['London Borough'].isin(nonBoroughs)]
NaNFreeDF2.head(-10)
```

```python
NaNFreeDF2['London Borough'].unique()
```

```python
df = NaNFreeDF2
df.head()
```

```python
df.dtypes
```

```python
# Visualizing the data of the borough of Camden
camden_prices = df[df['London Borough'] == 'Camden']

ax = camden_prices.plot(kind = 'line', x = 'Date', y = 'Average Price')
ax.set_ylabel('Price')
```

```python
df['Year'] = df['Date'].apply(lambda t: t.year)
df.tail()
```

```python
df[['London Borough', 'Year']].head()
```

```python
df[['London Borough', 'Year']].head()
dfg = df.groupby(by=['London Borough', 'Year']).mean()
dfg.sample(10)
```

```python
dfg.head()
```

```python
dfg = dfg.reset_index()
dfg.head()
```

```python
def create_price_ratio(d):
    y1998 = float(d['Average Price'][d['Year']==1998])
    y2018 = float(d['Average Price'][d['Year']==2018])
    ratio = [y1998/y2018]
    return ratio
```

```python
create_price_ratio(dfg[dfg['London Borough']=='Barking & Dagenham'])
```
